```python
pcostdata = pd.read_csv('COL City Info Percentage.csv', encoding='latin', na_values='n/a')
pcostdata = pcostdata.explode('Factor %')
pcostdata['Factor %'] = pcostdata['Factor %'].astype('float')
pcostdata
```

```python
sns.lineplot(x ="City", y = "Factor %", hue = "COL Indicator", data = pcostdata)

plt.show()
```

```python
compdata = pd.read_csv('COL Sea Comparison.csv', encoding='latin')
compdata = compdata.explode('Salary')
compdata['Salary'] = compdata['Salary'].astype('float')
compdata = compdata.explode('COL Salary')
compdata['COL Salary'] = compdata['COL Salary'].astype('float')
compdata
```

```python
ax = sns.barplot(x = "City", y = "COL Salary", hue = "Job", data=compdata, palette='flare')
ax2 = sns.lineplot(x = "City", y = "Salary", hue = "Job", data=compdata, palette='crest')
```

```python
dcostdata = pd.read_csv('COL City Info.csv', encoding='latin', na_values='n/a')
dcostdata = dcostdata.explode('$')
dcostdata['$'] = dcostdata['$'].astype('float')
dcostdata
```
